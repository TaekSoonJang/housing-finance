```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data.csv', usecols=[n for n in range(11)], thousands=',')
```

```python
df.head()
```

```python
df['date'] = df.apply(lambda r: '{}-{:02d}'.format(r[0], r[1]), axis=1)
```

```python
df.iloc[:, 2:11].max()
```

```python
xlabel = df.loc[:, 'date']
plt.figure(figsize=(30,10))
ymin, ymax = 0, 0
for i in range(2, 11):
    data = df.iloc[:, i]
    ymin = min(data.min(), ymin)
    ymax = max(data.max(), ymax)
    plt.plot(data)
plt.xticks(np.arange(len(xlabel)), labels=xlabel, rotation='vertical')
plt.yticks(np.arange(ymin, ymax, step=1000))
plt.grid(True)
plt.show()
```
